```python
import pandas as pd
file_path = "../data/dfly-scaling-data.xlsx"
df = pd.read_excel(file_path, sheetname='Sheet1')
print(df.shape)
df.drop_duplicates()
print(df.shape)
```

```python
df = pd.get_dummies(df)
df = df.loc[:, df.var() != 0.0]
```

```python
xtrain = df.drop(['runtime','ev_rolledback'], axis = 1)
features = xtrain.columns
```

```python
import scipy.stats
from matplotlib import pyplot as plt
import matplotlib as mpl
%matplotlib inline
import numpy as np

def visualize(y_label):
    fig, axs = plt.subplots(5,4, figsize=(20, 20), facecolor='w', edgecolor='k')
    fig.subplots_adjust(hspace = 1, wspace=1)

    axs = axs.ravel()

    for i in range(20):

        y = df[y_label]
        y2 = xtrain.iloc[:,i]
        axs[i].scatter(y,y2)
        axs[i].set_xlabel(features[i])
        axs[i].set_ylabel(y_label)

        PCC = round(scipy.stats.pearsonr(y,y2)[0],3)

        axs[i].text(np.min(y)*1.1, 0.95*np.max(y2), r'PCC='+str(PCC),
             fontsize=10)
```

```python
visualize("runtime")
```

```python
visualize("ev_rolledback")
```
